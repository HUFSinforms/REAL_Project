```python
import pandas as pd
import time,datetime
import numpy as np
import glob
from pandas import *
import scipy.optimize

#timeseries data
all_data=pd.read_csv('Timeseries_data_SP500.csv')
date_list=list(set(list(all_data['DATE'])))
for i in range(len(date_list)):
    ak=date_list[i]
    date_list[i]=ak[6:]+str('-')+ak[:2]+'-'+ak[3:5]
date_list.sort()
all_data['DATE']=pd.to_datetime(all_data['DATE'])
all_data=all_data.rename(columns={'NAME':'name','DATE':'date','SEDOL':'sedol','SECTOR':'sector','BETA':'beta','ALPHA_SCORE':'as','BENCH_WEIGHT':'bw',"MCAP_Q":'mq'})

dic_data = {k: v for k, v in all_data.groupby('date')}
```

```python
#risk:cov_mat
risk_data=pd.read_csv('cov_mat_2007-01-03.csv')
risk_sedol=risk_data['ROW_INDEX'].unique().tolist()
risk_mat = np.zeros((len(risk_sedol),len(risk_sedol)))
risk_mat[np.triu_indices(len(risk_sedol), 0)] = list(risk_data['VALUE'])
irows,icols = np.triu_indices(len(risk_sedol),0)
risk_mat[icols,irows]=risk_mat[irows,icols]
```

```python
import class_informs as inform
import pickle
import random
import heapq
import pp
import sys
import time

sam = 3
lambda_ = 30
ngen = 5
as_ = dic_data[Timestamp('2007-01-03 00:00:00')]['as']
dim = len(dic_data[Timestamp('2007-01-03 00:00:00')])
gen = 0

prob = [0]*dim
prob_car = []
r_a = []
population = []
selected = []

a0000 = []
eda_output = []
eda_obj = []

f2 = open('pic/risk_sedol.txt', 'rb')
sedol = pickle.load(f2)
ex = inform.informs(10000)
f8 = open('pic/risk_mat.txt', 'rb')
risk_mat = pickle.load(f8)
ex.set_omega(risk_mat)

ppservers = ()
if len(sys.argv) > 1:
    ncpus = 4
    job_server = pp.Server(ncpus, ppservers=ppservers)
else:
    job_server = pp.Server(ppservers=ppservers)
print ("Starting pp with", job_server.get_ncpus(), "workers")
    
def cp1(candidate, ex, sedol, dim) :
    ex.set_con(cons=candidate)
    new_list = []
    sol = ex.solve(0.01)[0]
    for k in sedol :
        new_list.append(sol[k])
    return new_list
#     try :
#         ex.set_con(cons=candidate)
#         new_list = []
#         sol = ex.solve(0.01)[0]
#         for k in sedol :
#             new_list.append(sol[k])
#         return new_list
#     except :
#         print("no solution")
#         car = int(random.random()*20+50)
#         candidate = random.sample(range(dim), car)
#         print(car)
#         cp1(candidate, ex, sedol, dim)
#         print("zzzzzzzzzzzaaaaaa")


def appe(result):
    population.append(result)

def generate():
    
    r_a = []
    candi__ = []
    
    for i in range(lambda_) :
        candi = []
        candi_ = []
        car = int(random.random()*20+50)
        if gen == 0 :
            candi = random.sample(range(dim), car)
        else :
            if i%2 ==1 :
                car = random.choice(prob_car)
            for j in range(dim) :
                if random.random() <= prob[j]/sam :
                    candi.append(j)
            if len(candi)>car :
                candi = random.sample(candi, car)
            elif len(candi) <car :
#                 large = heapq.nlargest(car, prob)
#                 candi_ = [prob.index(k) for k in large]
#                 candi += random.sample(set(candi_)-set(candi), car-len(candi))
                candi += random.sample(set(range(car))-set(candi), car-len(candi))
        candi__.append(candi)
        
    inputs = tuple(candi__)

    start_time1 = time.time()
    indsw=0
    for input in inputs :
        job_server.submit(cp1, (input, ex, sedol, dim, ), modules=("random", ), callback=appe)
        print("--- %s seconds submit in for ---" %(time.time() - start_time1))
        print(indsw)
        indsw += 1
    print("--- %s seconds pp submit end ---" %(time.time() - start_time1))    

    job_server.wait()
    print("--- %s seconds wait end ---" %(time.time() - start_time1))

    print(population)
    
    for po in population:
        print(po)
        if po != None:
            s_rr = np.dot(po, risk_mat)
            s_r = np.dot(s_rr, po)
            s_a = np.dot(po, as_)
            global r_a
            r_a.append(s_r-s_a)
    return population

        
def update(population) :
    selected = []
    global r_a
    small = heapq.nsmallest(sam, r_a)
    print('this is minimum object value')
    print(min(r_a))
    eda_output.append(population[r_a.index(min(r_a))])
    eda_obj.append(min(r_a))
    
    selected = [population[r_a.index(i)] for i in small]

    for i in range(len(selected)) :
        car_sel = 0
        for j in range(dim) :
            if selected[i] != None:
                if selected[i][j] > 0 :
                    prob[j] += 1
                    car_sel += 1
        prob_car.append(car_sel)
        
    global gen
    gen += 1
    population.clear()
```

```python
start_time = time.time()
for i in range(ngen) :
    population = generate()
    print("--- %s seconds generate---" %(time.time() - start_time))
    update(population)
    print("--- %s seconds update---" %(time.time() - start_time))
    if i == ngen-1 :
        f00 = open('output_weight.txt','wb')
        pickle.dump(eda_output, f00)
        f11 = open('output_objective.txt','wb')
        pickle.dump(eda_obj, f11)
```
