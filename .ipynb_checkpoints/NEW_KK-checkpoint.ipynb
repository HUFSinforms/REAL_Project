함수 정의

```python
def cal_obj(sol, risk_mat, alpha) :
    s_rr = np.dot(sol, risk_mat)
    s_r = np.dot(s_rr, sol)
    s_a = np.dot(sol, alpha)
    obj = s_r+s_a
    return obj*10000

def select_candi(sol) :
    diff_sub = np.array([0]*len(alpha), float)
    diff_add = np.array([0]*len(alpha), float)
    candi = []
    
    pre = cal_obj(sol, risk_mat, alpha)
    
    for i in range(len(sol)) :
        if(sol[i] != 0) :
            w = sol[i]
            sol[i] = 0
            diff_sub[i] = cal_obj(sol, risk_mat, alpha)
            if min(diff_sub[diff_sub!=0]) == diff_sub[i] :
                for j in range(len(sol)) :
                    if sol[j] == 0 :
                        w1 = sol[j]
                        sol[j] = w
                        diff_add[j] = cal_obj(sol, risk_mat, alpha)
                        sol[j] = w1
            sol[i] = w
            
    sol[numpy.where(diff_sub == max(diff_sub[diff_sub!=0]))[0][0]] = 0
    sol[numpy.where(diff_add == max(diff_add[diff_add!=0]))[0][0]] = 0.1
    
    for val in sol :
        if val > 0 :
            candi.append(sol.index(val))
            
    return candi
```
